2)Load the dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/House_Price_India[1].csv')
df.head()
```

3)Perform visualisations

```python
#univariate analysis
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(df['number of bedrooms'], bins=20, kde=True)
plt.title('Number of Bedrooms Distribution')

plt.subplot(2, 2, 2)
sns.histplot(df['number of bathrooms'], bins=20, kde=True)
plt.title('Number of Bathrooms Distribution')

plt.subplot(2, 2, 3)
sns.histplot(df['living area'], bins=20, kde=True)
plt.title('Living Area Distribution')

plt.subplot(2, 2, 4)
sns.histplot(df['Price'], bins=20, kde=True)
plt.title('Price Distribution')

plt.tight_layout()
plt.show()
```

```python
#bivariate analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='living area', y='Price', hue='number of bedrooms')
plt.title('Relationship between Living Area, Price, and Number of Bedrooms')
plt.show()
```

```python
#multivariate analysis
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#pairplots
sns.pairplot(data=df, vars=['living area', 'lot area', 'Price', 'number of bedrooms'], hue='number of bathrooms')
plt.title('Pair Plot for Living Area, Lot Area, Price, and Number of Bedrooms (Colored by Number of Bathrooms)')
plt.show()
```
